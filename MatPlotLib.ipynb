# MatPlotLib Labels and Title

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([4, 20])
y = np.array([2, 18])
plt.xlabel("x axis")
plt.ylabel("y axis")


plt.plot(x) 
plt.plot(y)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([4, 20])
y = np.array([2, 18])
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("Ogechukwu Stephanie Oboka")

plt.plot(x) 
plt.plot(y)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([4, 20])
y = np.array([2, 18])

font1 = {"family":"serif", "color": "green", "size": 20}
font2 = {"family":"serif", "color": "blue", "size": 10}
plt.xlabel("x axis", fontdict = font2)
plt.ylabel("y axis", fontdict = font2)
plt.title("Ogechukwu Stephanie Oboka", fontdict = font1)

plt.plot(x) 
plt.plot(y)
plt.show()
```
