## Marker Colour

```python
import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([10, 20, 3, 14])

plt.plot(ypoints, marker = "d", ms = 40, mec = "y") #outlines the marker with the specified colour
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([10, 20, 3, 14])

plt.plot(ypoints, marker = "D", ms = 40, mfc = "g") #fills the marker with the specified colour
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([10, 20, 3, 14])

plt.plot(ypoints, marker = "d", ms = 25, mec = "y", mfc = "y") #changes the colour of both the inside and outside
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([10, 20, 3, 14])

plt.plot(ypoints, marker = "o", ms = 25, mec = "g", mfc = "y") #changes the colour of both the inside and outside
plt.show()
```
